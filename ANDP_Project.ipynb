```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the data
df = pd.read_excel('sales_data.xlsx')
```

```python
df.head()
```

```python
df.tail()
```

```python
#Basic structure and data types
df.info()
```

```python
#Summary statistics for numerical data
df.describe()
```

```python
#Check for missing values
df.isnull().sum()
```

```python
#Count unique values in categorical columns
df.nunique()
```

```python
#Convert Sale_Date to datetime objects
df['Sale_Date'] = pd.to_datetime(df['Sale_Date'])
print(df['Sale_Date'].dtype)
```

```python
#Extract Year and Month for time-based analysis
df['Year'] = df['Sale_Date'].dt.year
df['Month'] = df['Sale_Date'].dt.month_name()
df[['Sale_Date', 'Year', 'Month']].head()
```

```python
#Check for and remove duplicate rows
duplicate_count = df.duplicated().sum()
df = df.drop_duplicates()
print(f"Removed {duplicate_count} duplicates.")
```

```python
#Calculate Gross Profit per transaction
df['Profit'] = (df['Unit_Price'] - df['Unit_Cost']) * df['Quantity_Sold']
df[['Product_ID', 'Sales_Amount', 'Profit']].head()
```

```python
#Calculate Profit Margin percentage
df['Margin_Percentage'] = (df['Profit'] / df['Sales_Amount']) * 100
df['Margin_Percentage'].mean()
```

```python
#Calculate Discount Value in currency
df['Discount_Amount'] = df['Sales_Amount'] * df['Discount']
df[['Sales_Amount', 'Discount', 'Discount_Amount']].head()
```

```python
#Total Sales by Region
df.groupby('Region')['Sales_Amount'].sum().sort_values(ascending=False)
```

```python
#Top 5 Sales Representatives by Total Profit
df.groupby('Sales_Rep')['Profit'].sum().nlargest(5)
```

```python
#Average Sales Amount per Region
df.groupby('Region')['Sales_Amount'].mean().plot(kind='bar', color='skyblue')
plt.title('Average Sale Value by Region')
plt.show()
```

```python
#Sales Representative contribution within each Region
df.groupby(['Region', 'Sales_Rep'])['Sales_Amount'].sum().unstack().fillna(0)
```

```python
#Total Quantity Sold per Product Category
df.groupby('Product_Category')['Quantity_Sold'].sum().sort_values(ascending=False)
```

```python
#Most profitable Product Categories
sns.boxplot(x='Product_Category', y='Profit', data=df)
plt.title('Profit Distribution by Category')
plt.show()
```

```python
#Identify the Top 10 most expensive products
df[['Product_ID', 'Product_Category', 'Unit_Price']].drop_duplicates().nlargest(10, 'Unit_Price')
```

```python
#Compare Average Discount vs Category
df.groupby('Product_Category')['Discount'].mean().sort_values().plot(kind='barh')
plt.title('Average Discount Rate by Category')
plt.show()
```

```python
#Sales Distribution: Online vs Retail
df.groupby('Sales_Channel')['Sales_Amount'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('')
plt.title('Sales Channel Revenue Share')
plt.show()
```

```python
#Revenue from New vs Returning Customers
df.groupby('Customer_Type')['Sales_Amount'].sum()
```

```python
#Preferred Payment Methods
sns.countplot(data=df, x='Payment_Method', palette='viridis')
plt.title('Frequency of Payment Methods')
plt.show()
```

```python
#Average spending: New vs Returning Customers
df.groupby('Customer_Type')['Sales_Amount'].mean()
```

```python
#Monthly Sales Trend (Total Revenue)
df.groupby('Month')['Sales_Amount'].sum().reindex([
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
]).plot(kind='line', marker='o')
plt.title('Monthly Sales Trend 2023')
plt.show()
```

```python
#Identify the busiest Sales Dates
df['Sale_Date'].value_counts().head(10)
```

```python
#Average Daily Sales across the week
df['Day_of_Week'] = df['Sale_Date'].dt.day_name()
df.groupby('Day_of_Week')['Sales_Amount'].mean().reindex([
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
]).plot(kind='bar')
plt.show()
```

```python
#Correlation Matrix (Numeric Columns only)
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Distribution of Sales Amount (Histogram)
sns.histplot(df['Sales_Amount'], kde=True, bins=30)
plt.title('Sales Amount Distribution')
plt.show()
```

```python
#Relationship between Quantity and Profit
sns.scatterplot(x='Quantity_Sold', y='Profit', data=df, hue='Product_Category')
plt.title('Quantity Sold vs Profit')
plt.show()
```

```python
#Pivot Table: Region vs Payment Method (Transaction Count)
df.pivot_table(index='Region', columns='Payment_Method', values='Sales_Amount', aggfunc='count')
```

```python
#Filter transactions with high discounts (> 20%)
high_discount_sales = df[df['Discount'] > 0.20]
high_discount_sales.head()
```

```python
#Grouped summary for reporting
summary = df.groupby(['Region', 'Product_Category']).agg({
    'Sales_Amount': 'sum',
    'Profit': 'sum',
    'Quantity_Sold': 'mean'
}).reset_index()
summary
```
